```python
import league
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

summonerFrame = pd.DataFrame(columns = ["summonerName", "id", "puuid", "summonerLevel", "tier", "rank", "wins", "losses", "leaguePoints"])

# Get league data
summonerFrame.loc[len(summonerFrame)] = league.getSummonerData("Funeral Home")
summonerFrame.loc[len(summonerFrame)] = league.getSummonerData("Bushido ßrown")
summonerFrame.loc[len(summonerFrame)] = league.getSummonerData("Best URGOT in NA")
summonerFrame.loc[len(summonerFrame)] = league.getSummonerData("Voldemort")
summonerFrame.loc[len(summonerFrame)] = league.getSummonerData("NeftLut")

summaryFrame = summonerFrame.drop(columns = ["id", "puuid"])
summaryFrame["w/r"] = (summaryFrame["wins"] / (summaryFrame["wins"] + summaryFrame["losses"])) * 100
summaryFrame["w/r"] = summaryFrame["w/r"].map('{:.2f}%'.format)
summaryFrame
```
